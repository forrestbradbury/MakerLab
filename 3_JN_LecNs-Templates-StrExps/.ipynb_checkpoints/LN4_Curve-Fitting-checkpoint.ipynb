<span style="font-size:larger;color:brown">**Curve fitting**</span>

Curve fitting consists of finding parameters from some self-chosen type of function that make that function best agree with some measured data.  Specifically, the data will consist of some measured variable (called the "dependent variable") that is paired to and plotted against some independent variable (or variables).  

The function type used in curve-fitting is usually one that our theories predict will explain a relationship between the dependent and independent variables; but occasionally, the function type is just chosen because it seems to nicely fit the empirical data (a useful model that doesn't have a theoretical justification is called a "phenomenological model").

There are several different Python libraries that assist with curve fitting.  The `scipy` library is one of the more popular, simple, and flexible ones.  However, a newer and more specialized library called `lmfit` is even more user friendly.  Code examples for using both are given below, with the latter being the recommended option.

Please learn by example, going through the following codes, and trying out your own curve-fitting skills in the given exercises.

# Fully-worked example:

```python
#  before curve-fitting, we import useful libraries, load data from a file, & plot

import math
import numpy as np
import matplotlib.pylab as plt
from pathlib import Path  # pathlib.Path formats pathnames to suit your operating system

# "notebook" plotting mode allows for interactive (zoom-able) plots, "inline" doesn't
# %matplotlib notebook   
%matplotlib inline


# declare the appropriate filename as a string:
filename = "periodicdata1.txt"   # was sent in zipfile along with this Jupyter Notebook

sub_folder = Path("datafolder/")     # set name of sub-directory where data is to be stored
file_path = sub_folder / filename    # file path is combination of filename and its folder

array = np.loadtxt(file_path, delimiter='\t') # loads file's data (tab-separated) into "array"
                                              # for comma-separated-value data:  delimiter=','
    
# "array" has 2 columns:  first for times, second for signal values
# we transpose (.T) and assign the 0th and 1th rows as individual 1D arrays:
timearray, signalarray = array.T
    
plt.figure(figsize=(8, 4))
plt.plot(timearray, signalarray, ".") 
plt.xlabel("time (seconds)")
plt.ylabel("signal (arbitrary units)")
plt.show()

#  NOTE:  if you're in %matplotlib notebook Mode, you may have to run this twice before it works
```

Awesome!  We made a plot of the file's data!  Here, the signal column is plotted against the time column, and thus we are treating the signal as the dependent variable as we plot it against the changing independent variable: time.

But, it looks like these data represent two very different conditions (one before oscillations have started, and one after).  Let's assume they represent the position of a swinging pendulum, and the initial values correspond to the situation before the pendulum has been released.  Thus, let's assume these **initial data points are USELESS and we want to get rid of them in our further analysis!**

If we use the plot-interactions mode chosen by the `%matplotlib notebook` command, we can zoom into the region where the oscillations appear to start (and see that this occurs at about 8 seconds on the x-axis).  To be very safe in knowing that have have definitely entered the oscillating regime, we will truncate the dataset and remove all data points occuring before $time = 8.1$ seconds:

```python
## this code creates new time and signal arrays with the useless data removed

##  THESE CODE LINES ARE FOR REMOVING DATA FROM BEGINING
## this is our estimate of independent variable's value when useful signal starts
cut_until = 8.1  # in this example, this is time=8.1 seconds
## searchsorted() function finds the timearray index before which our "cut_until" value
##   would fit between the ordered timearray elements
##   https://numpy.org/doc/stable/reference/generated/numpy.searchsorted.html
cut_until_index = timearray.searchsorted(cut_until)

## THESE CODE LINES ARE FOR REMOVING DATA FROM END (which we don't do in this example)
## if there is useless data at the end, we set a cut_after to delete it too: 
cut_after = 35  # we give a value larger than max time, just to show it still works
cut_after_index = timearray.searchsorted(cut_after)

## HERE WE ACTUALLY TRUNCATE BOTH OUR 1D ARRAYS
## this [i:j] selects only values from index "i" until index "j"
cuttimearray    =  timearray[cut_until_index:cut_after_index]
## because time and signal points are paired, we must truncate signalarray identically:
cutsignalarray  =  signalarray[cut_until_index:cut_after_index]

plt.figure(figsize=(8, 4))
plt.plot(cuttimearray, cutsignalarray, ".")
plt.xlabel("time (seconds)")
plt.ylabel("signal (arbitrary units)")
plt.show()
```

The figure above confirms that we have cut out the data values corresponding to signals and times before oscillations had started.  Now we can try fitting this curve with an appropriate function, $signal(t)$, which models the signal's dependence on time.  

Knowing that pendula oscillate roughly sinusoidally and that any resistance in the system will slowly decrease the oscillation amplitude over time, we might think that an exponentially decaying sinusoid is a good model for our data:

$$
signal(t) = A e^{-k t} \sin{\left(\omega t+\phi\right)}
$$

where, $A$ is the initial amplitude (at time $t=0$), $k$ is the exponential decay rate, $\omega$ is the angular frequency (which is in radians per second and thus relates to the regular frequency as:  $\omega = 2\pi f$), and $\phi$ is the phase shift with respect to the normal sine curve.

We now have a data set and a mathematical model ("curve" or "function") with which we'd like to fit it.  There are now options for how to accomplish this in Python:

- the `scipy` library's `optimize` functions 
- the `lmfit` library's `models` functions  (recommended)

Here below, we include the `scipy.optimize` method in raw text form, and the `lmfit.models` in Python code cells because we think you will find the latter method easier.

Now we can use `lmfit.models` for curve-fitting:


```python
## lmfit documentation:  https://lmfit.github.io/lmfit-py/model.html
from lmfit import models

## here we define the function (aka curve) that we're fitting the data with:
##    it's an exponentially decaying sinusoid with arbitrary amplitude and phase
def test_func(x, A, omega, k, phi):
    return A * np.exp(-k*x) * np.sin(omega * x + phi)  

model = models.Model(test_func)  # we declare our fitting curve to be the test_func
guess = model.make_params(A=10, omega=5, k=.1, phi=0)  # initial guesses for parameter values
fit = model.fit(cutsignalarray, x=cuttimearray, params=guess)  # performs the fit!
    # "fit()"  -> (dep var)   , x=(indep var) , parameters' initial guesses

## printing fitted parameter values from a special dictionary created by model.fit()
print(fit.params.values())
## assigning names for the fitted parameter values
A_fit = fit.params['A'].value           # fitted amplitude 
omega_fit = fit.params['omega'].value   # fitted frequency (in rad/s) 
k_fit = fit.params['k'].value           # fitted decay rate 
phi_fit = fit.params['phi'].value       # fitted phase offset

## this plots 1. the fit's "residuals" and 2. the data versus fitted curve:
fit.plot(datafmt=',')  # the comma yields very small lines and data points 
plt.xlabel("time (seconds)")
plt.ylabel("signal (arbitrary units)")

## here we convert and print the fitted frequency in Hertz
freq_fit = omega_fit / (2 * math.pi)
print(f"The estimated frequency via fitting with decaying sinusoid: {freq} Hz")
```

Above, we already see one of the nice features of the `lmfit.models` output.  The fit's simple `.plot()` command automatically plots the measured & fitted data AND above it gives a plot of **residuals**.  

These "residuals" are the differences between the dependent variable of actual measured data and the curve-fit's estimation.  In this specific dataset, there just appears to be some constant Gaussian-distributed noise on top of the exponentially decaying oscillations.  In your own experiments, you might instead see noise that grows with signal magnitude or possibly some extra functional dependence which was not introduced by your `test_func`.  Thus, the residuals plot can often help you understand your sources of experimental uncertainty and/or the limitations of your theoretical (mathematical) model.  

While you could construct your own plot of residuals from the `scipy.optimize` code results above (by plotting the difference: `cutsignalarray - test_func(...)` against `cuttimearray`), it's really nice that the `lmfit.models` code automatically does it for us :)

# Standard uncertainty of fitted parameters

In a previous Lecture Notes document, we learned about uncertainties and uncertainty propagation.  Here, we must realize that the parameter values found via curve-fitting also have uncertainty!  

A curve fit is optimized according to the available data, but since those data will have some natural variability, we must try to understand the uncertainty of our fitted parameters in their ability to predict experimental outcomes.  Since these uncertainties relate to the variability of measured data, they are a kind of **Type A uncertainty**.

For both of the methods below (using either the `scipy.optimize` or the `lmfit.models` functions), we will make use of the `uncertainties` package, which may need installed if you haven't done so already:


## <span style="color:gray">Parameter uncertainties with `scipy.optimize`</span>

<span style="font-size:larger;color:red">**NOTE AGAIN:  the `lmfit.models` method below is usually recommended over this one!**</span>

<span style="color:gray">The `optimize.curve_fit()` function actually already calculates what we need to find the uncertainty in fitted parameter values:
`fitted_params, covariance_matrix = optimize.curve_fit( INPUTS... )`
The second output is a square two-dimensional array called the covariance matrix:
https://en.wikipedia.org/wiki/Covariance_matrix</span>

<span style="color:gray">For the example above, the covariance matrix yields a matrix with 4 rows and 4 columns:</span>

<span style="color:gray">To keep things simple, we will ignore the off-diagonal terms in this diagonal matrix (called the co-variances between each of the four fitted parameters), and focus instead on the diagonal terms which equal the square of the standard uncertainties of each of the four parameters (in other words, the standard variances of those parameters):</span>

<span style="color:gray">Finally, if we wanted to determine the periodicity of this data set, we should propagate the standard uncertainty in angular frequency ($u_{\omega}$) to the standard uncertainty in period ($u_T$):</span>

<span style="color:gray">Knowing that:  $T\ = \ \frac{1}{f} \ = \ \frac{2\pi}{\omega}$</span>

<span style="color:gray">we have:  $u_T \ = \ \sqrt{\left( \frac{\partial T}{\partial \omega} u_{\omega} \right)^2} \ = \ \left| \frac{\partial T}{\partial \omega} u_{\omega} \right| \ = \ \frac{2\pi}{\omega ^2} u_{\omega}$</span>



<span style="color:gray">Instead of doing the algebra ourselves for propagating the standard uncertainty of $\omega$ to $u_T$, we can alternatively use the `uncertainties` package:</span>

<span style="color:gray">And, now let's see how the more specialized curve-fitting library accomplishes the same tasks:</span> 

## Parameter uncertainties with `lmfit.models`

The `model.fit()` function is even easier, as it creates an object called `fit` which contains all parameter values with uncertainties and the fit's reduced chi-squared statistic ($\chi^2_\text{red}$) which ought to be $\approx 1$ for a well-chosen and well-fitted mathematical model.

```python
fit
```

Most importantly for us are the **parameter uncertainties (refered to above as the "standard errors")**, as they are the type A uncertainties estimated by the variability in our data set.

To work with these parameter values and uncertainties further, we can assign them convenient names and use the uncertainties package.  Here, we print all parameter values to the appropriate number of significant figures and then propagate the uncertainty in angular frequency through to the oscillation period:

```python
##  First we assign simpler names to each of the parameter uncertainties:

u_A = fit.params['A'].stderr  
u_omega = fit.params['omega'].stderr  
u_k = fit.params['k'].stderr  
u_phi = fit.params['phi'].stderr  


## here we report estimate & uncertainty using the convenient uncertainties package:

from uncertainties import ufloat
# from uncertainties.umath import *  # would be needed for sin(), etc.

Ampl = ufloat(A_fit, u_A)  # = estimate +- standard uncertainty
Angfreq = ufloat(omega_fit, u_omega)  # = estimate +- standard uncertainty
Decayrate = ufloat(k_fit, u_k)  # = estimate +- standard uncertainty
Phase = ufloat(phi_fit, u_phi)  # = estimate +- standard uncertainty

## this (.2u) prints the correct number of significant figures in the uncertainty
print(f'Estimation of amplitude from our curve-fit, and its standard uncertainty:         {Ampl:.2u} seconds')
print(f'Estimation of angular frequency from our curve-fit, and its standard uncertainty: {Angfreq:.2u} rad/s')
print(f'Estimation of decay rate from our curve-fit, and its standard uncertainty:        {Decayrate:.2u} 1/s')
print(f'Estimation of phase offset from our curve-fit, and its standard uncertainty:      {Phase:.2u} rad')
print('\n')  #print blank line

T = 2*math.pi / Angfreq

## this (.2u) prints the correct number of significant figures in the uncertainty 
print(f'Estimation of period from our curve-fit, and its standard uncertainty:            {T:.2u} seconds')
```

And, that's all!  The long example above takes us through the selection of data relevant for curve-fitting, the use of a curve-fitting function to estimate parameter values, and the method of finding Type A uncertainties in these estimated parameter values.

# Linearization

Instead of comparing fitted-curves, we can sometimes change the plotted dependent variable for quicker evaluation of a possible functional relationship.  For instance, if we hypothesize that the world population is growing exponentially in time, then:  $pop = A e^{kt}$, and taking a natural log of both sides we obtain: $\log_e {pop} = (\log_e {A}) + kt$, and thus one could model the log of the population as a straight line, $y=mx+b$ corresponding to: $\log_e {pop} = mt + b$ where $b=\log_e {A}$ and $m=k$.

If you analyzed the world population data in the previous code cell, you will have already seen that an exponential curve doesn't fit the data perfectly (as it over-predicts how fast the population has recently been growing).  

Here below, we look at the urban population growth.  The following codes compare the regular population plot with plots of $\log_e {pop}$ as well as $\sqrt{pop}$, with the latter performed in case a quadratic function (a second order polynomial) is the best fit.  By visually comparing which scatter plot is best approximated by a straight line (and is thus *linearized*), we make a quick conclusion about which function best describes how urban population has been increasing in time.

## <span style="color:gray">Linearization and fitting with `scipy.optimize`</span>

<span style="font-size:larger;color:red">**NOTE AGAIN:  the `lmfit.models` method below is usually recommended over this one!**</span>


```python
####  USING `scipy.optimize`  ####

# here, we choose to study the urban population growth:
# the previous code cell already loaded data for:  yeararray & urbanpoparray 

# our test function for all plots is a simple linear one:
def linear_test_func(x, m, b):
    return m*x+b  

# optimize.curve_fit needs as input: the function, data arrays & initial parameter guesses
#    it outputs an estimation of the function's parameter values 
#               and a "covariance matrix"
regparams, regparams_covariance = optimize.curve_fit(linear_test_func, yeararray, 
                                            urbanpoparray, p0=[1,1])  # p0= initial guesses

logurbanpop = np.log(urbanpoparray)
logparams, logparams_covariance = optimize.curve_fit(linear_test_func, yeararray, 
                                            logurbanpop, p0=[1,1])  # p0= initial guesses

sqrturbanpop = np.sqrt(urbanpoparray)
sqrtparams, sqrtparams_covariance = optimize.curve_fit(linear_test_func, yeararray, 
                                            sqrturbanpop, p0=[1,1])  # p0= initial guesses

plt.figure(figsize=(12, 5))

plt.subplot(1,3,1)  # 1row, 3columns, plotindex=1
plt.plot(yeararray,urbanpoparray,'.k', label='Data')
plt.plot(yeararray, linear_test_func(yeararray, regparams[0], regparams[1]), label='Fitted function')
plt.xlabel('year')
plt.ylabel('world urban population')  
plt.title('regular population plot')
plt.legend(loc='best')

plt.subplot(1,3,2)  # 1row, 3columns, plotindex=2
plt.plot(yeararray,logurbanpop,'.k', label='Data')
plt.plot(yeararray, linear_test_func(yeararray, logparams[0], logparams[1]), label='Fitted function')
plt.xlabel('year')
plt.ylabel('log(world-urban-population)')  
plt.title('plot of log of data')
plt.legend(loc='best')

plt.subplot(1,3,3)  # 1row, 3columns, plotindex=2
plt.plot(yeararray,sqrturbanpop,'.k', label='Data')
plt.plot(yeararray, linear_test_func(yeararray, sqrtparams[0], sqrtparams[1]), label='Fitted function')
plt.xlabel('year')
plt.ylabel('sqrt(world-urban-population)')  
plt.title('plot of square root of data')
plt.legend(loc='best')

plt.tight_layout()   # prevents plots and axis labels & titles from overlapping

# Display the figure:
plt.show()
```

The first plot includes the linear fit of the regular data series and shows it's got some nonzero (and positive) curvature.  This demonstrating that the world's urban population is not just growing, but its speed of growth has been getting bigger.  

However, the log of the data (which would be linearized if they were exponentially growing) is fit slightly less well by a straight line (here, showing that the annual percentage growth does not remain constant - like for exponential growth, but rather has decreased over the past 20 years).  So, the population as a function of time has been accelerating, but not according to an exponential curve.  

The last plot of the square root of urban population seems to be a bit more linear than the first two.  While we have no theory saying that the recent world urban population growth ought to be a quadratic function of time, we could conclude that such a second order polynomial is an OK phenomenological explanation.  However the plot of residuals show that our linear best fit of the square root data under-estimates the real values, then over-estimates, then under-estimates them again as time goes on...

The fact that the world's urban population cannot be easily fit by such simple mathematical models should not surprise us.  Luckily, many phenomena in nature can be sufficiently isolated from environmental noise and display these more basic mathematical relationships - which can be demonstrated via linearizations like these.
