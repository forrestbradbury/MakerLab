<span style="font-size:24pt">**Fourier transforms**</span>


Fourier's theorem is very useful in several types of mathematical analyses.  It states that any periodic function can be written as the sum of sinusoidal functions (sines or cosines) of different amplitudes, frequencies, and phases.

The most familiar example is that a Fourier transform can convert a time-domain sound signal (experimentally corresponding to measured pressure induced over time at a microphone) into its spectrum of tones.  In other words, a Fourier transform decomposes a sound signal into amplitudes of its different tones (frequencies)!

Here, we mostly spare you all the mathematical details.  But, if you're interested, please have a look online for some nice materials:

# Discrete Fourier Transform

There are different types of Fourier transforms, and the terminology is confusing.  When working with measured data, a certain type of Fourier transform is needed:  one that acts on real-valued data (they are measured, so they are not imaginary!) which are taken at discrete and constantly spaced time steps.  The transform which works with discrete data is sometimes refered to as the discrete Fourier transform, but sometimes rather called a fast Fourier transform (fft).  We will use one in Python's numpy library called `np.fft.rfft()` which is good for processing our real measurements.

A description of our Fourier transform can be found here:
https://docs.scipy.org/doc/numpy/reference/generated/numpy.fft.rfft.html

Further, since the Fourier transform yields complex-valued amplitudes of each frequency value and we often only care about the magnitudes of those frequency contributions and not their phases, then we can immediately take the absolute value (or modulus) of the Fourier output:  `np.absolute( np.fft.rfft(dataarray) )`

A fully worked example of using this particular Fourier transform follows below.


******

## First, data are loaded and the oscillating region is selected:


````python
#  before Fourier transforming, we import libraries, load data from file & plot

import math
import numpy as np
import matplotlib.pylab as plt
from pathlib import Path  # pathlib.Path formats pathnames to suit your operating system

# "notebook" plotting mode allows for interactive (zoom-able) plots, "inline" doesn't
# %matplotlib notebook   
%matplotlib inline


# declare the appropriate filename as a string:
filename = "periodicdata1.txt"   # was sent in zipfile along with this Jupyter Notebook

sub_folder = Path("datafolder/")     # set name of sub-directory where data is to be stored
file_path = sub_folder / filename    # file path is combination of filename and its folder

array = np.loadtxt(file_path, delimiter='\t') # loads file's data (tab-separated) into "array"
                                              # for comma-separated-value data:  delimiter=','
    
# "array" has 2 columns:  first for times, second for signal values
# we transpose (.T) and assign the 0th and 1th rows as individual 1D arrays:
timearray, signalarray = array.T
    
plt.figure(figsize=(8, 4))
plt.plot(timearray, signalarray, ".") 
plt.xlabel("time (seconds)")
plt.ylabel("signal (arbitrary units)")
plt.show()

#  NOTE:  if you're in %matplotlib notebook Mode, you may have to run this twice before it works
````

The code's output above should show a plot of the file's data.  Here, the signal column is plotted against the time column, and thus we are treating the signal as the dependent variable as we plot it against the changing independent variable: time.

**Note:  since we hope to Fourier transform the oscillating part of this signal, it's worthwhile to double-check by zooming into this graph (or just inspecting the source-datafile) that the data points are indeed evenly spaced in time.  Remember that variable time-spacing would prevent a Fourier transform from giving valid results!**

In this case, the data points are evenly spaced in time, but it looks like these data represent two very different conditions (one before oscillations have started, and one after).  Let's assume they represent the position of a swinging pendulum, and the initial values correspond to the situation before the pendulum has been released.  Thus, let's assume these **initial data points are USELESS and we want to get rid of them in our further analysis!**

If we use the plot-interactions mode chosen by the `%matplotlib notebook` command, we can zoom into the region where the oscillations seem to start (and see that this occurs at about 8 seconds on the x-axis).  To be very safe in knowing that have have definitely entered the oscillating regime, we will truncate the dataset and remove all data points occuring before $time = 8.1$ seconds:


````python
## this code creates new time and signal arrays with the useless data removed

##  THESE CODE LINES ARE FOR REMOVING DATA FROM BEGINING
## this is our estimate of independent variable's value when useful signal starts
cut_until = 8.1  # in this example, this is time=8.1 seconds
## searchsorted() function finds the timearray index before which our "cut_until" value
##   would fit between the ordered timearray elements
##   https://numpy.org/doc/stable/reference/generated/numpy.searchsorted.html
cut_until_index = timearray.searchsorted(cut_until)

## THESE CODE LINES ARE FOR REMOVING DATA FROM END (which we don't do in this example)
## if there is useless data at the end, we set a cut_after to delete it too: 
cut_after = 35  # we give a value larger than max time, just to show it still works
cut_after_index = timearray.searchsorted(cut_after)

## HERE WE ACTUALLY TRUNCATE BOTH OUR 1D ARRAYS
## this [i:j] selects only values from index "i" until index "j"
cuttimearray    =  timearray[cut_until_index:cut_after_index]
## because time and signal points are paired, we must truncate signalarray identically:
cutsignalarray  =  signalarray[cut_until_index:cut_after_index]

plt.figure(figsize=(8, 4))
plt.plot(cuttimearray, cutsignalarray, ".")
plt.xlabel("time (seconds)")
plt.ylabel("signal (arbitrary units)")
plt.show()
````

***

##  Fourier transform and plotting

Now, we are ready to Fourier transform the above data:

As you see below, using the `np.fft.rfft()` function is easy, but relating its output to real frequencies is a bit trickier.  After calculating them, the code plots the signal in four different ways.

>The following six lines of code do all the necessary calculations, while the rest of the below code cell consists of comments and plotting commands.  
>```   
fftsignalarray = np.absolute( np.fft.rfft(cutsignalarray) )    
n = len(cuttimearray)    
timesteps = n - 1       
dt = ( cuttimearray[-1:] - cuttimearray[0] ) / ( timesteps )   
fPs = 1 + (n // 2)    
freqarray = np.arange( fPs ) / (len(cuttimearray)*dt)  
```
>The FFT data consist of the frequencies (`freqarray`) and their amplitudes (`fftsignalarray`).


<span style="color:red">*NOTE:  the code below assumes the array's time data are given in seconds, and thus yields frequencies in Hz.*</span>


<span style="color:red">*NOTE ALSO:  the code's creation of the frequency axis is explained in more detail in this document's last section below.*</span>

````python
#  This program plots the time-domain signal, 
#  and then Fourier transforms and replots in frequency-domain
#
#  NOTE:  the Fourier Transform REQUIRES that signal measurements:
#         1. have equal time spacing
#         AND
#         2. are arranged in chronological order in the array 
#
#  NOTE ALSO:  we assume time data is in seconds, and thus give frequencies in Hz

# Use discrete Fourier transform for real input:   numpy.fft.rfft()
# https://docs.scipy.org/doc/numpy/reference/generated/numpy.fft.rfft.html
fftsignalarray = np.absolute( np.fft.rfft(cutsignalarray) )  
#  the Fourier transform yields both a magnitude & phase of each frequency contribution
#  it does so with real & imaginary components: 
#                            remember Euler's formula:  e^iwt = cos(wt) + i sin(wt)  
#  numpy.absolute() takes modulus of complex numbers which describe frequency contributions  
#  this gives us strength of the frequency's contribution to signal and ignores its phase


## NOTE:  FREQUENCY AXIS EXPLAINED IN LAST SECTION BELOW
#
# to plot strengths of frequenies' contributions, we need associated list of frequencies:
# what is the time step between points, "dt" ? 
n = len(cuttimearray)  # number of data points in arrays
timesteps = n - 1      # n-1 time steps occur between n points
dt = ( cuttimearray[-1:] - cuttimearray[0] ) / ( timesteps )   #  (max - min)/steps 
#  note: [-1:] is counting backwards from end, yielding time array's very last value 
#
# which frequencies correspond to Fourier transform's output data points?
# when n=even, max frequency = (sampling rate/2) , period = 2*dt, measuring hi/lo/hi/lo...
# smallest non-zero frequency = 1/(n*dt) = has one period per measurement time window
# next smallest non-zero frequency = 2/(n*dt) = has two periods per measurement time window
# etc...   
# (and similar for n=odd, see above-mentioned website)
#
# NOTE: besides smallest non-zero frequency, also "zero frequency" for overall signal offset
#
# to get number of freq Points (fPs), including an extra for the 'zero frequency' term:
fPs = 1 + (n // 2)  # "n//2" throws out remainder keeping only integer for odd n
freqarray = np.arange( fPs ) / (len(cuttimearray)*dt)  # arange(i) yields [0,1,2,3,...,i-1]


###  NOTE:  often a dataset's "zero frequency" amplitude is meaningless because it just 
###         gives the constant signal offset instead of showing any periodic behavior
## to delete the zero frequency amplitude, uncomment these two lines:
#fftsignalarray = fftsignalarray[1:]   # [1:] selects all but the 0th element
#freqarray = freqarray[1:]             # [1:] selects all but the 0th element


# now we are ready to plot Fourier spectrum: 
# (fftsignalarray = contribution strengths) VS (freqarray = their respective frequencies)

plt.figure(figsize=(7, 10))

#First, for comparison we plot the signal in the time domain
plt.subplot(4,1,1)
plt.plot(cuttimearray,cutsignalarray,'.k')  #plot with black (k) dots (.)
plt.xlabel('time (seconds)')
plt.ylabel('signal \n (arbitrary units)')  
plt.title('time domain plot')

#Here, we plot the signal in the frequency domain (aka the Fourier spectrum)
plt.subplot(4,1,2)
plt.plot(freqarray,fftsignalarray,'k*')
plt.xlabel('frequency (Hz)')
plt.ylabel('amplitudes \n (arbitrary units)')  
plt.title('frequency domain plot')

#repeat frequency domain plot, but zoomed in on x-axis near dominant frequency
plt.subplot(4,1,3)
plt.plot(freqarray,fftsignalarray,'k*')
plt.xlim(0, 1.6)          #  sets min and max x-axis limit around peak of ~0.8Hz
plt.xlabel('frequency (Hz)')
plt.ylabel('amplitudes \n (arbitrary units)')  
plt.title('frequency domain plot, zoomed to dominant frequency (not automatic!)')

#Here, we plot the log of the amplitudes versus log of their frequencies
plt.subplot(4,1,4)
if freqarray[0]==0:
    plt.plot(np.log(freqarray[1:]),np.log(fftsignalarray[1:]),'k*')  
    # note:  [1:] skips the zero frequency point because log0 = undefined
else:        
    plt.plot(np.log(freqarray),np.log(fftsignalarray),'k*')  
    # note:  else the zero frequency point had already been deleted above!

plt.xlabel('log frequency')
plt.ylabel('log amplitudes \n (arbitrary units)')  
plt.title('frequency domain log-log plot')

plt.tight_layout()   # prevents plots and axis labels & titles from overlapping

# Display the figure:
plt.show()
````

NOTE:  The example dataset used here did not have a large zero frequency peak because the data were centered about zero.  If you use this code for Fourier transforming data which has an offset, you will probably wish to uncomment the lines above which remove the zero frequency and its amplitude:

> ```  
fftsignalarray = fftsignalarray[1:]      # [1:] selects all but the 0th element
freqarray = freqarray[1:]                # [1:] selects all but the 0th element
```

##  Analyzing the Fourier spectrum

In the second plot generated by the code above, we see a small Fourier peak corresponding to higher frequency noise (100 Hz) on the signal, but it is clear that most of the signal is explained by a sinusoidal oscillation near 0.8 Hz because that is by far the largest peak in the Fourier frequency spectrum.

To better see this dominant peak, we can either zoom into the second plot (assuming we have already switched on the plot interaction mode with: `%matplotlib notebook`), or we can replot with an already restricted x-axis like in the third plot, or we can even plot the log of the x & y axes like in the fourth plot.  

When zoomed in, we see the discretization of the FFT data:  it's not a continuous curve, but rather has discrete data points.  The FFT takes in discrete and evenly-spaced data and subsequently produces amplitudes for a discrete and evenly spaced set of frequencies.  This discretization prevents us from very precisely reading off a signal's "true" oscillation frequency(ies).  

Without taking any more data, we could improve our estimate of the frequency of a peak by fitting a quadratic or Gaussian peak to the data points around the FFT peak, with the frequency at the maximum of the fitted curve giving us a better estimate for the signal's dominant frequency(ies) than does the maximum of the discrete Fourier spectrum.  For more such methods of estimating peak positions from a discrete Fourier spectrum, see:  https://dspguru.com/dsp/howtos/how-to-interpolate-fft-peak/  Please note, however, that these peak approximation methods are NOT more accurate than finding a frequency by directly curve-fitting with a sinusoid (which is demonstrated in the curve-fitting Jupyter Notebook).  But, one advantage of the Fourier transform is that it always works and - contrary to curve-fitting - doesn't depend on good initial guesses of parameter values.  Another important advantage is that it finds other oscillatory signals hidden within the dominant signal (like the 100 Hz noise we observe in this data set).

In case you find it useful, a quick algorithm for estimating the peak frequency by using a quadratic fit of only the highest and two neighboring data points is given below:

*Note that broad and noisy Fourier peaks that stretch over many frequencies are better approximated with other algorithms, such as a Gaussian mentioned above.*

````python
#  quadratic approximation method 
#  see:  https://dspguru.com/dsp/howtos/how-to-interpolate-fft-peak/

index4max = np.argmax(fftsignalarray)  # argmax() gives the index of the max of the array
print("the max frequency data point is at:  " + str(freqarray[index4max])+" Hz")

y1 = fftsignalarray[index4max-1]  # y1= point just left of max
y2 = fftsignalarray[index4max]    # y2= max amplitude
y3 = fftsignalarray[index4max+1]  # y3= point just right of max

# instead of using a fancy curve-fitting package, 
# quadratic curve-fitting of just 3 data points can be exactly solved to find:
# this fractional shift in x-axis variable for the estimated max frequency
fractionalindex = (y3 - y1) / (2 * (2 * y2 - y1 - y3))  
# note that a quadratic function has three fitting parameters,
# and when only three data points are used in the fit, the relation is solved exactly
# without the need for a least squares regression!, yielding the "fractionalindex" above
freqstep = freqarray[index4max+1] - freqarray[index4max]
diff = fractionalindex * freqstep
est_peak_freq = diff+freqarray[index4max]
print("the estimated true max frequency is:  " + str(est_peak_freq)+" Hz")



#  here, we go further by plotting the estimated peak alongside the Fourier transform data

est_peak_ampl = y2 + (2*y2 - y1 - y2)/2 * fractionalindex**2

plt.figure(figsize=(6, 4))
#First, we plot the signal in the time domain

plt.plot(freqarray,fftsignalarray,'k.')   # black dots
plt.plot(est_peak_freq,est_peak_ampl,'r*')  # red star
plt.xlim(0.65, .95)          #  sets min and max x-axis limit around peak of ~0.8Hz
plt.xlabel('frequency (Hz)')
plt.ylabel('amplitudes \n (arbitrary units)')  
plt.title('Fourier transform & quadratic fit near peak')
plt.show()

````

# Min & Max Fourier frequencies

Besides the "zero frequency" amplitude which was discussed above, it's important to know how your measured data determine the Fourier spectrum's range of frequencies.  Specifically, let's consider the minimum frequency (aka the frequency step size) and the maximum frequency (aka the "Nyquist frequency").

***For the purpose of this discussion, and just like in the example above, we will refer to the measured data's independent variable as "time", however please keep in mind that the Fourier transform can also analyze periodic behavior dependent on other independent variables, such as spatially periodic oscillations.***

Because frequency and period are inversely related, it should not come as a surprise that the measured data's total range in time determines the Fourier spectrum's smallest frequency increment, whereas the measured data's smallest time increment determines the Fourier spectrum's total range in frequency.



## Frequency step size

As depicted below in Figure 11, the periodic signal with the smallest frequency which can be derived from the measured data will complete one full oscillation in the total measurement time.  The next smallest frequencies complete two, three, ... full oscillations, leading to frequency increments $\left(\Delta f\right)$ equal to this smallest (non-zero) frequency.  Thus:
$$
\Delta f\ =\ f_{min} \ = \ \frac{1}{T_{max}} \ = \ \frac{1}{time_{_{total.measurement}}}
$$
where the minimum frequency has a maximum period $T_{max}$ equal to the total measurement time.


## Maximum (Nyquist) frequency

Also depicted below in Figure 11 is the periodic signal with the largest frequency which can be derived from the measured data.  Because amplitude determination requires observing both a high and a low signal value in each oscillation period, the maximum frequency which can be derived from the measured data will complete one oscillation in a time equal to two of the measured data's time steps.
$$
f_{max} \ = \ \frac{1}{T_{min}} \ = \ \frac{1}{2 \Delta t}
$$
where the maximum frequency has a minimum period $T_{min}$.  This maximum frequency is also known as the **Nyquist frequency**, and we see that it is equal to half the "sampling rate" of our measured data:
$$
f_{Nyquist} \ = \ \frac{sampling.rate}{2} \ = \ \frac{1/\Delta t}{2}
$$

## Importance for experimental design

The above discussion makes clear that how often and for how long we measure determines, respectively, the minimum and maximum frequencies of the Fourier transformed data:

| time axis (of input)|               | frequency axis (of output)|
|:----------------    |:------------- |:---------------------     |
| total time range    | *determines:* |  frequency increment     |
| time increment     | *determines:* |  total frequency range    |

The fact that the total measured time range determines the frequency increment means that:  For slow oscillations (low frequencies) or for very precise frequency resolution, we need to measure for a longer time.

The fact that the measuring time increment determines the total frequency range means that:  For fast oscillations (high frequencies), we need to measure with smaller time steps (increments).  If our measurand is oscillating at frequencies near to or greater than our sampling frequency, we will miss the effects we are trying to observe and instead observe the confusing effect of "aliasing":  https://en.wikipedia.org/wiki/Aliasing
