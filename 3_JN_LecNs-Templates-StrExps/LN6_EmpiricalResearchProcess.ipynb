<span style="font-size:28pt">**Background**</span>


## Experimental skills

Experimental research skills can be usefully categorized into the cognitive tasks conducted by experimental researchers.  An open-inquiry science project helps train these skills by practicing all the parts of the empirical research process.


## Iteration

The tasks (and sub-tasks) listed below are not simply chronological steps.  Rather, as Figure 1 indicates, researchers must iteratively improve and loop back as more insights are made and more data are collected which point to superior experimental methods.  For example, conducting and analyzing results from initial exploratory tests or experiments will help to inform, refine, and improve your research goals and subsequent measurements.

## Challenge of open inquiry

Many of tasks in the figure above are difficult.  However, the real challenge is knowing when to cycle back and when to move forward.  The freedom of designing your own open-inquiry project comes with the responsibility of successfully navigating the empirical research process!

# Establish research goals

What are the goal(s) and question(s) of your research project?  The following sub-tasks will help you decide whether they are suitable:


## 	Assess level of research question

Is your research question at the appropriate level?  

*For this course your project should involve measuring scientific phenomena and/or utilizing advanced methods which are beyond what you learned in high school.*

## 	Review relevant literature

A literature review involves collecting theoretical models and experimental methods & results.  

*A short experiment in a Bachelor's level lab course has negligible chance of significantly extending the boundaries of scientific knowledge, thus an extensive literature review is not necessary (nor possible given the time constraints).  But, your results must be compared to relevant scientific models and empirical studies, and knowing of them ahead of time is critical for guiding your experimental process.* 

## 	Initial feasibility assessment

Initial evaluation of whether the project topic, goals, and research question are consistent with the constraints on time, equipment, and expertise.

## 	Initial safety assessment

Conduct initial safety assessment to flag possible issues (see section 9).


******
# 	Define criteria for suitable evidence

What scientific evidence could be used to conclusively answer your research question?  

The following sub-tasks help in deciding what will constitute suitable evidence to achieve the goal by developing criteria and/or utilizing existent ones.

## 	What types of measured data would be convincing?  

Note that typically there are many sorts of measurements that can provide evidence for answering any given question.

## 	Important variables

What variables are important and how might they be measured and controlled?

## 	Experimental controls

What types of experimental controls and checks would need to be in place?


*******
# 	Determine feasibility of experiment

After identifying what kinds of measurements would be required for suitable scientific evidence, you can better assess the feasibility of answering your research question.  This and the previous task require some iteration before moving onto the specifics of experimental design. 

## 	Predict feasibility

Is it realistically possible to carry out the experiment?  If so, are the required time and resources available and reasonable?  (This involves a more detailed reiteration of 1.3.)

## Contingency options

As a researcher, you must also analyze contingency options, in case the results of the experiment are not what you hoped for.  Will the data produced still provide interesting information which can be reported?  Will the results show how to improve the apparatus to achieve conditions needed to obtain hoped-for results?

## Expanded Safety Assessment

Revisit and expand the safety assessment (see section 9).


********
# 	Experimental design

Your experimental design must focus on answering the research question, while being practical and safe.

## 	Design details

You should explore many possible preliminary designs (the basic design has likely already been identified in previous steps, but now its details should be carefully considered).  These design details will need revisited based on your considerations in the sub-tasks below, and then again after initial tests have been conducted.

### 	Physical materials & devices and their usage

Experimental methods must specify the actual materials and devices and the basics of their usage (more detailed experimental procedures follow later). 

### 	Data analysis methods

You must also specify methods for data processing and analysis.  The method by which your experimental data are compared with theoretical models influences the uncertainties (see next sub-task).

## 	Identify uncertainties and limits of models/assumptions

Identify sources of measurement uncertainty and analyze relevant variables that may lead to systematic errors in results and interpretation.  What are the effects of imperfect models and assumptions?  Quantifying these requires having complex cause and effect models for the experiment.   (This sub-task will be repeated after measuring performance of the apparatus.)

## Revisit Safety Assessment & adapt accordingly

Adapt plans according to your safety assessment.


## Optimize design

Finalize the design, taking into account construction details and performance requirements of each component.  

## Data acquisition strategy

Experimental design also includes data-taking procedures.  These must take into account constraints on time, clarity of results, all potential statistical and systematic uncertainties, and the importance and requirements for distinguishing between different potential interpretations of results.  An uncertainty budget which estimates the effects of different (propagated) sources of uncertainty will help in deciding how to prioritize your time & effort.  For instance, you may decide that some unimportant quantities need not be measured.  (These sub-tasks will be repeated after measuring performance of the apparatus.)

### Set parameters and range of independent variable(s)

Both the range and step-size of the independent variable are important for your experiment's sensitivity to the expected effects, as too small of a range means too small of an effect while too large of a step-size (and too few data points) prevents a reliable comparison to a theoretical model.  

The choices for fixed parameters are also often important for maximizing effect size, minimizing uncertainty, or simplifying experimental design.

### Sequence of measurements

The sequences that measurements are taken can play two important roles.  

First, many physical systems exhibit **hysteresis**.  When the effects of hysteresis cannot be eliminated, they can instead be worked around by controlling the order that measurements are taken.

Second, sometimes the desired ranges of independent variables are not known in advance and must be sorted out during an experiment.  Starting with "safe" and/or "interesting" values and analyzing on the fly allows researchers to efficiently adjust their measurement plans accordingly.

### Data averaging

Data averaging is important whenever type A uncertainties are a significant part of your uncertainty budget.  Repetition of measurements and longer signal averaging increase precision and accuracy by decreasing the effects of uncontrolled & variable sources of noise.  Unfortunately, even the simplest sources of type A uncertainty only decrease by a factor of $1/\sqrt{N}$ or $1/\sqrt{T}$, where $N$ and $T$ are the number or measurements and the averaging time, respectively.  Thus, the need to measure for 100x longer to achieve a factor of 10 decrease in such a type A uncertainty may limit your potential improvements.

# 	Construct and test apparatus and implement procedure

While you should have already played around with your experimental equipment to figure out how it works and what it does, now things get serious!

## 	Schedule of construction

Decide who should build the various parts and on what schedule.  Remember to leave plenty of time for further iteration.  Spending all your time on the first attempt means that no time is left to improve on it!

## Design tests and/or calibrations of individual components

Develop criteria and test procedures for evaluation of the apparatus components.  

## Test individual components and the fully assembled setup

Test individual components as they are completed and iterate earlier tasks when the test results are different than expected.

Collect data on performance of individual components and full apparatus.  These will be useful in assessing the possible measurement ranges & parameter values and in uncertainty propagation calculations.  It is also critically important for finding and fixing any malfunctions.

## Modify components or setup

The test results often point to useful ways of modifying particular components or the overall setup.


## Revisit experimental design

Reiterate task 4.5, taking into account actual performance of finished apparatus.


## Collect experimental data

Remember here the importance of sequencing (task 4.5.2).

Record your observations (because you never know which uncontrolled variables may turn out to be important).

#  Analyze data

Take the advice of other students who have completed open-inquiry projects and start early with your data analysis!  If data cannot be analyzed on the fly, then analyze them as soon as possible thereafter.  This is especially important because your initial results usually suggest important ways to improve the experimental design and data acquisition.


## Find suitable mathematical models

A theoretical/mathematical model is always just an approximation of a real physical situation.  In your data analysis, you will have to decide which approximations are justified and which are not, and under which conditions.


## Statistically analyze data

Decide on and implement appropriate statistical analysis methods and procedures, and apply them to data from all relevant measurements (where the dismissal of outliers or irrelevant measurements must be clearly motivated).  

You must be able to argue why your analysis is suitable to your data set.  For instance, statistical tests for significance can show the likelihood of a set of measurements under a given null hypothesis - but they often implicitly assume that type B systematic uncertainties are negligible.  In this case it is better to propagate the type B uncertainty through to the bounds of a confidence interval (itself constructed from the statistical uncertainty).  Thus, the next two sub-tasks can either justify or refute your methods in this one.



## Calculate type A (statistical) uncertainty

For a set of simple repeated measurements, the type A uncertainty is equal to the experimental standard deviation of the mean.  For the determination of a curve fitting parameter through measurements with changing independent variables, the type A uncertainty is equal to the square root of that parameter's diagonal element in the covariance matrix.


## Final calculation (propagation) of all type B measurement uncertainties

You have probably already done this for previous tasks, so this is one final check that the calculations and assumptions are still valid.

# Evaluate results and analyze their implications

Drawing conclusions from your experimental data involves translating a collection of numbers into solid scientific evidence.


## Comparing results to expectations

In what ways do your results agree with expectations?  In what ways are they different?

### If different, why?

When results come out differently than expected, reasons should be sought.  

#### Relationships and pattern recognition

This involves calling on complex mental models incorporating a web of cause and effect relationships, strategies for separating relevant and irrelevant (or insignificant) information, and complex pattern recognition.  

#### Alternative models or un-accounted-for factors

Consider plausible interpretations or new theoretical or experimental directions which are motivated by your results?  Can these new ideas be further tested by additional analyses of the existing data or can the experimental methods be adjusted to better evaluate them?  Are there other variables that should be measured and controlled?

####  Test new hypotheses

In order to test new hypotheses, it is usually necessary to conduct extra measurements suited to these new questions.  This will often involve modification of apparatus and reiterating many of the previous tasks.


### If they agree, rigorous confirmation

What are the limits of this agreement?  How can you exclude experimenter biases?  Humans often repeat experiments when results are different than expected, but then fail to do so when initial results meet expectations, leading to a bias towards confirmation.  

Perform extra measurements and redundant tests to confirm your observations, reduce type A statistical uncertainties, and estimate the confounding effects of systematic type B uncertainties.  Repeat previous tasks as needed.

# Presenting the work

The communication of an open scientific inquiry requires serious effort, as no-one else has performed your experiments before.  It is completely up to you to explain your methods and results!


## Communicating methods

Your experimental setup and data analysis methods must be introduced and motivated.  While you should focus on the final or best iterations, you should also briefly explain why those choices were made.


## Communicating results

Follow standard data display procedures or, as needed, develop new procedures that highlight critical features of your results.


## Communicating the significance of results

Explain the work so the broader context and uniqueness of the work, the apparatus, the procedures, and the conclusions are easily understood, and the audience/readers perceive it to be of maximum interest and significance.

Focus on sharing what you know and why you understand it.  There may still be some unsolved mysteries and room for improvement, but your audience must first understand the purposes and successes of your experiments.

# Notes on Safety Assessments and Ethical Considerations

It is probable that no-one has ever performed exactly the same experiments that you will do for your open-inquiry project.  When sailing into uncharted waters, there can be risks!  

**Safety and ethical concerns include data privacy issues and/or potential risks to persons, animals, property, or the natural environment.**

Regarding safety, in addition to any concerns regarding the scientific equipment, you must also consider issues relating to using (or mis-using) common household objects.  Reviewing user manuals for correct and safe usage of all equipment (scientific or otherwise) is mandatory.  

Also, if and when you adjust your experimental methods during the course of this project, any changes relating to safety or ethical/privacy issues should be first checked with instructors before implementation.


<span style="font-size:larger;color:red">**Please know that ethical concerns or potentially unsafe practices are a valid excuse for any of the project team members to halt your experimental procedures at any time, even if your instructors did not previously recognize the concerns.  Since you will be conducting experiments of your own design outside of a supervised setting, the final responsibilty lies in your hands.**</span>
